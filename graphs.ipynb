```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 200)
y = np.vectorize(sigmoid)(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
def relu(x):
    return max(0.0, x)
```

```python
y = np.vectorize(relu)(x)
```

```python
plt.plot(x, y)
```

```python
def plot(func, start = -10, end = 10):
    x = np.linspace(start, end, 100)
    y = np.vectorize(func)(x)
    plt.plot(x, y)
```

```python
plot(sigmoid, -1, 1)
```

```python
plot(sigmoid, -10, 10)
```

```python
plot(relu, -100, 100)
```

```python
plot(sigmoid, -100, 100)
```
